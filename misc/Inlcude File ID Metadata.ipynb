```python
from os.path import splitext
import pandas as pd
from dateutil import parser
```

```python
df = pd.read_csv("file_IDs.csv")
df
```

```python
df["tile"] = df.granule_ID.apply(lambda granule_ID: granule_ID.split("_")[2])
df
```

```python
df["date"] = df.granule_ID.apply(lambda granule_ID: parser.parse(granule_ID.split("_")[3]).date())
```

```python
df["variable"] = df.filename.apply(lambda filename: splitext(filename)[0].split("_")[-1])
```

```python
df.variable
```

```python
df.to_csv("google_drive_file_IDs.csv", index=False)
```
